```python
from fastai.tabular.all import *
```

```python
data_path = 'C:/Users/light/Documents/Fastbook/Personal-Machine-Learning-Data/'
```

```python
df = pd.read_csv(data_path + 'simple calculator.csv')
```

```python
df.head()
```

```python
splits = RandomSplitter(valid_pct=0.1)(range_of(df))
```

```python
to = TabularPandas(df, procs=[Categorify],
                  cont_names = ['num1', 'num2'],
                  cat_names = ['symbol'],
                  y_names = 'answer',
                  y_block=CategoryBlock,
                  splits=splits)
```

```python
len(to.train),len(to.valid)
```

```python
to.xs.iloc[:2]
```

```python
dls = to.dataloaders(bs=64, num_works = 0)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
```

```python
learn.fit_one_cycle(100)
```

```python
learn.show_results()
```

```python
d = {'num1': [2], 'symbol': ['add'], 'num2': [5]}
df2 = pd.DataFrame(data=d)
df2
```

```python
row, clas, probs = learn.predict(df2.iloc[0])
```
