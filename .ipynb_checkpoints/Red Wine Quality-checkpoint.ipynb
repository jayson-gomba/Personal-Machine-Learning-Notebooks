```python
def rf(xs, y, n_estimators=40, min_samples_leaf=48, **kwargs):
    return RandomForestClassifier(n_jobs=-1, 
                                  n_estimators=n_estimators, 
                                  min_samples_leaf=min_samples_leaf, 
                                  oob_score=True).fit(xs, y)
```

```python
m = rf(xs, y);
```

```python
m.score(xs, y), m.score(valid_xs, valid_y)
```

```python
def rf_feat_importance(m, df):
    return pd.DataFrame({'cols':df.columns, 'imp':m.feature_importances_}).sort_values('imp', ascending=False)
```

```python
fi = rf_feat_importance(m, xs)
fi
```

```python
def plot_fi(fi):
    return fi.plot('cols', 'imp', 'barh', figsize=(12, 7), legend=False)
```

```python
df_nn = pd.read_csv('winequality-red.csv', sep=';', low_memory=False)
```

```python
df_nn_final = df_nn[list(xs_final_t.columns) + [dep_var]]
```

```python
df_nn_final.columns
```

```python
xs_filt2 = xs_filt
valid_xs_t2 = valid_xs_t
m2 = rf(xs_filt2, y_filt)
m2.score(xs_filt2, y_filt), m2.score(valid_xs_t2, valid_y)
```

```python
splits_nn = RandomSplitter(valid_pct=0.2)(range_of(df_nn_final))
```

```python
cont_nn = ['alcohol', 'sulphates', 'volatile acidity', 'total sulfur dioxide',
       'density']
```

```python
to_nn = TabularPandas(df_nn_final, 
                      procs=[Categorify], 
                      cont_names=cont_nn, 
                      y_names=dep_var, 
                      splits=splits_nn, 
                      y_block=CategoryBlock)
```

```python
dls = to_nn.dataloaders(1024)
```

```python
y = to_nn.train.y
y.min(), y.max()
```

```python
learn = tabular_learner(dls, y_range=(0, 5), layers=[500,250], metrics=accuracy)
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(40)
```

```python
preds,targs = learn.get_preds()
accuracy_score(preds.argmax(axis=-1), targs)
```

```python
learn.save('redwine_nn')
```
